# Imports

```python
import os
import shutil
```

# Dataset Multiplicator

```python
# Ustaw ścieżkę do głównego folderu
base_folder = r'C:\Users\Michał\Documents\ISWD\workdir_doubled\beer_labels'

# Definiujemy rozszerzenia, które uznajemy za obrazy
allowed_exts = {'.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff', '.tif'}

# Iterujemy przez wszystkie elementy w folderze głównym
for subfolder in os.listdir(base_folder):
    subfolder_path = os.path.join(base_folder, subfolder)
    
    # Sprawdzamy, czy dany element jest podfolderem
    if os.path.isdir(subfolder_path):
        # Wyszukujemy pliki obrazkowe, które nie zawierają "_copy" w nazwie
        images = [
            f for f in os.listdir(subfolder_path)
            if os.path.isfile(os.path.join(subfolder_path, f)) 
            and os.path.splitext(f)[1].lower() in allowed_exts 
            and '_copy' not in os.path.splitext(f)[0]
        ]
        
        # Jeśli podfolder nie zawiera żadnych obrazów, pomijamy go
        if not images:
            # print(f"Folder {subfolder} jest pusty lub nie zawiera obrazów – pomijam.")
            continue

        # Dla każdego obrazu tworzymy kopię
        for image in images:
            original_path = os.path.join(subfolder_path, image)
            name, ext = os.path.splitext(image)
            new_filename = name + "_copy" + ext
            new_path = os.path.join(subfolder_path, new_filename)
            
            # Jeśli plik z nową nazwą już istnieje, dopisujemy licznik, aby uniknąć nadpisania
            counter = 1
            while os.path.exists(new_path):
                new_filename = f"{name}_copy{counter}{ext}"
                new_path = os.path.join(subfolder_path, new_filename)
                counter += 1
            
            shutil.copy2(original_path, new_path)
            print(f"Skopiowano: {original_path} -> {new_path}")
```
